```python
import pickle
```

```python
with open("nrpooledvalues_acc_entropy.pkl", 'rb') as file:
    nrpooledvalues_acc_entropy = pickle.load(file)
print("Pickle file loaded successfully.")

nrpooledvalues_acc_entropy
```

```python
with open("nrpooledvalues_acc_random.pkl", 'rb') as file:
    nrpooledvalues_acc_random = pickle.load(file)
print("Pickle file loaded successfully.")

nrpooledvalues_acc_random
```

```python
with open("nrpooledvalues_acc_dropout_diff.pkl", 'rb') as file:
    nrpooledvalues_acc_dropout_diff = pickle.load(file)
print("Pickle file loaded successfully.")

nrpooledvalues_acc_dropout_diff
```

```python
import matplotlib.pyplot as plt
```

```python

x1 = list(nrpooledvalues_acc_entropy.keys())
y1 = [max(nrpooledvalues_acc_entropy[key]) for key in x1]

x2 = list(nrpooledvalues_acc_random.keys())
y2 = [max(nrpooledvalues_acc_random[key]) for key in x2]

x3 = list(nrpooledvalues_acc_dropout_diff.keys())
y3 = [max(nrpooledvalues_acc_dropout_diff[key]) for key in x3]

plt.plot(x1, y1, label='Entropy')
plt.plot(x2, y2, label='Random')
plt.plot(x3, y3, label='Dropout diff')
plt.xlabel('Number of labeled Datapoints')
plt.ylabel('Accuracy')
# plt.title('Accuracy vs. Number of labeled Datapoints')
plt.legend()
plt.show()
```
